Este notebook utilizará o dataset Sandstone, disponibilizado pela Elsevier, utilizado no livro  <a href="https://booksite.elsevier.com/9780444537966/?ISBN=9780444537966">Physical Properties of Rocks</a>, exemplo do livro texto de Darling (2005) para demonstrar análises de testemunho e de log.

As variáveis analisadas incluem:

- **Depth**: profundidade  
- **GR**: raio gama  
- **RHOB**: densidade  
- **NPHI**: porosidade neutrônica  
- **RES\_DEEP**: resistividade profunda

Esse dataset é bastante útil para estudos de interpretação petrofísica, análise estatística e correlação entre propriedades físicas das rochas.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Carregando dataset sandstone
df = pd.read_excel('../data/csv/sandstone.xlsx')
df.head()
```

```python
# Descrição estatística do dataset
df.describe()
```

```python
# Plotando os perfis geofísicos

cols = ['GR','RES-DEEP','RHOB','NPHI']
unit = ['API','$\Omega \cdot$m','g/cm$^3$','a.a']

fig,axes = plt.subplots(ncols=4,figsize=(9,6))

aux=0

for ax,c,u in zip(axes,cols,unit):
    if aux!=1:
        ax.set_title(c,fontsize=16)
        ax.set_xlabel(u,fontsize=12)
        ax.invert_yaxis()
        ax.plot(df[c],df.DEPTH)
    if aux==1:
        ax.set_title(c,fontsize=16)
        ax.set_xlabel(u,fontsize=12)
        ax.invert_yaxis()
        ax.plot(df[c],df.DEPTH)
        ax.set_xscale('log')
    if (aux>0):
        ax.set_yticks([])
    else:
        ax.set_ylabel('Profundidade (m)',fontsize=14)
    aux+=1

fig.tight_layout()

fig.show()
```

```python
#Criando um gráfico de dispersão entre as variáveis GR e RHOB
fig, ax = plt.subplots(figsize=(7,5))


ax.scatter(df['GR'], df['RES-DEEP'], edgecolor='black')
ax.set_yscale('log')
ax.grid(True, which='both', linestyle='--', linewidth=0.5)

ax.set_title('Gráfico de Dispersão', fontsize=16)
ax.set_xlabel('GR (API)', fontsize=14)
ax.set_ylabel(r'Resistividade $\Omega \cdot$m', fontsize=14)
```

```python
fig, ax = plt.subplots(figsize=(7,5))

# Plotando pontos com GR < 60
ax.scatter(
    df[df.GR < 60]['GR'],
    df[df.GR < 60]['RES-DEEP'],
    c='orange',
    edgecolor='black',
    label='GR < 60'
)

# Plotando pontos com GR > 60
ax.scatter(
    df[df.GR > 60]['GR'],
    df[df.GR > 60]['RES-DEEP'],
    c='green',
    edgecolor='black',
    label='GR > 60'
)

ax.set_yscale('log')
ax.grid(True, which='both', linestyle='--', linewidth=0.5)

ax.set_title('Gráfico de Dispersão', fontsize=16)
ax.set_xlabel('GR (API)', fontsize=14)
ax.set_ylabel(r'Resistividade $\Omega \cdot$m', fontsize=14)

ax.legend()
```

Também é possível representar os dados em um **"pseudo-3D"**, no qual a coloração de cada ponto do gráfico de dispersão é associada ao valor de uma terceira variável. Assim, além das duas variáveis representadas nos eixos do gráfico, a **cor dos pontos** pode indicar o valor de uma propriedade física adicional.

```python
fig, ax = plt.subplots(figsize=(7,5))

sc = ax.scatter(
    df['GR'],
    df['RES-DEEP'],
    edgecolor='black',
    c=df['RHOB']
)

ax.set_yscale('log')
ax.grid(True, which='both', linestyle='--', linewidth=0.5)

cbar = fig.colorbar(sc, ax=ax)
cbar.set_label('Densidade', fontsize=14)

ax.set_title('Gráfico de Dispersão', fontsize=16)
ax.set_xlabel('GR (API)', fontsize=14)
ax.set_ylabel(r'Resistividade $\Omega \cdot$m', fontsize=14)
```

### Bubble Plot

O **gráfico de bolhas** é uma variação do gráfico de dispersão. Nesse tipo de visualização, as amostras são representadas por bolhas, cujo **tamanho** pode expressar uma dimensão adicional (ou uma nova propriedade).

```python
fig, ax = plt.subplots(figsize=(7,5))

sc = ax.scatter(
    df['GR'],
    df['RES-DEEP'],
    s=df['NPHI']*1000,
    edgecolor='black',
    c=df['RHOB'],
    label=r'$\phi_N$'
)

ax.set_yscale('log')
ax.grid(True, which='both', linestyle='--', linewidth=0.5)

# colorbar
cbar = fig.colorbar(sc, ax=ax)
cbar.set_label('Densidade', fontsize=14)

# títulos e rótulos
ax.set_title('Gráfico de Dispersão', fontsize=16)
ax.set_xlabel('GR (API)', fontsize=14)
ax.set_ylabel(r'Resistividade $\Omega \cdot$m', fontsize=14)

ax.legend()

plt.show()
```

### Pairplot

O **pairplot** é uma ferramenta de visualização utilizada para explorar relações entre múltiplas variáveis de um conjunto de dados. Esse tipo de gráfico apresenta, de forma simultânea, **gráficos de dispersão para cada par de variáveis**, permitindo identificar correlações, agrupamentos ou tendências entre elas.

Além disso, na diagonal da matriz são geralmente exibidos **histogramas ou distribuições** individuais de cada variável, facilitando a análise da distribuição dos dados.

```python
# Calculando o logaritmo da resistividade
df['log_RES'] = np.log10(df['RES-DEEP'])
df.head()
```

```python
# Criando um pairplot para visualizar as relações entre as variáveis
# A variável depth foi excluída

p_plot = sns.pairplot(
    df,
    vars=df.columns[1:],
    diag_kws={'edgecolor':'black'}
)

p_plot.fig.set_size_inches(8,8)
```

### Heatmap

O **heatmap** é uma técnica de visualização que utiliza **cores para representar a magnitude de valores em uma matriz de dados**. Cada célula da matriz é colorida de acordo com o valor que representa, permitindo identificar padrões, tendências e relações entre variáveis de forma rápida.

Em análises exploratórias de dados, o heatmap é frequentemente utilizado para visualizar **matrizes de correlação**, facilitando a identificação de relações positivas ou negativas entre diferentes propriedades.

```python
cov_matrix = df[df.columns[1:]].cov()
cov_matrix
```

```python
fig,axes = plt.subplots(figsize=(6,5))
sns.heatmap(cov_matrix,cmap='viridis',ax=axes,annot=True)
axes.set_title('Matriz de Covariância',fontsize=16)
fig.tight_layout()
```

```python
corr_matrix = df[df.columns[1:]].corr()
corr_matrix
```

```python
corr_matrix.columns
```

```python
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))



fig, ax = plt.subplots(figsize=(7,6))

sns.heatmap(
    corr_matrix,
    mask=mask,
    annot=True,          # ← mostra os valores
    fmt=".2f",
    cmap="viridis",
    square=True,
    linewidths=0.5,
    ax=ax
)

ax.set_title('Matriz de Correlação', fontsize=16)
```
